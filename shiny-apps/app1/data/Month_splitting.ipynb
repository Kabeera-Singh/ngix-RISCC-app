```python
import re
from calendar import month_abbr
import pandas as pd

def parse_months(bloom_period):
    if pd.isna(bloom_period) or bloom_period in ['Non-flowering', 'Indeterminate', 'Rarely flowers to Rarely flowers']:
        return []
    
    # Convert full month names to abbreviations
    for full_name, abbr in zip(['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December'],
                               ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']):
        bloom_period = bloom_period.replace(full_name, abbr)
    
    # Handle special cases
    if 'To frost' in bloom_period:
        end_month = 'Oct'  # Assuming frost typically occurs around October
        bloom_period = bloom_period.replace('To frost', end_month)
    
    months = re.findall(r'\b(Jan|Feb|Mar|Apr|May|Jun|Jul|Aug|Sep|Oct|Nov|Dec)\b', bloom_period)
    
    if '–' in bloom_period or 'to' in bloom_period.lower():
        start, end = months[0], months[-1]
        all_months = list(month_abbr)[1:]  # All month abbreviations
        start_index = all_months.index(start)
        end_index = all_months.index(end)
        if start_index <= end_index:
            months = all_months[start_index:end_index+1]
        else:
            months = all_months[start_index:] + all_months[:end_index+1]
    
    return months

# The input list
bloom_periods = df['BloomPeriod'].to_list()
# Process each entry
result = [parse_months(period) for period in bloom_periods]

# Print the result
for i, months in enumerate(result):
    print(f"{bloom_periods[i]}: {months}")
```
